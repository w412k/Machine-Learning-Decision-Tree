```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df.info()
```

```python
df.describe().T
```

```python
mask = df.isnull()
total = mask.sum()
percent = 100*mask.mean()

missing_data = pd.concat([total, percent], axis=1,join='outer',
                         keys=['count_missing', 'perc_missing'])
missing_data.sort_values(by='perc_missing', ascending=False, inplace=True)
missing_data
```

```python
import math 
median_age = math.floor(df.Age.median())
median_age
```

```python
df.Age.fillna(median_age)
```

```python
df.Age = df.Age.fillna(median_age)
df
```

```python
mask = df.isnull()
total = mask.sum()
percent = 100*mask.mean()

missing_data = pd.concat([total, percent], axis=1,join='outer',
                         keys=['count_missing', 'perc_missing'])
missing_data.sort_values(by='perc_missing', ascending=False, inplace=True)
missing_data
```

```python
ndf = df.drop(['Passenger', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked' ], axis ='columns')
ndf
```

```python
dummies = pd.get_dummies(ndf['Sex'])
dummies = dummies.astype(int)
dummies
```

```python
merged = pd.concat([ndf, dummies], axis = 'columns')
merged
```

```python
final = merged.drop(['Sex','female'], axis = 'columns')
final
```

```python
inputs = final.drop('Survived' , axis = 'columns')
target = final['Survived']
```

```python
from sklearn.model_selection import train_test_split

inputs_train, inputs_test, target_train, target_test = train_test_split(inputs,target,test_size=0.2) 
```

```python
inputs.head()
```

```python
target.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_train, target_train)
```

```python
model.score(inputs_test, target_test)
```

```python
model.predict([[3, 27, 7 ,1]])
```

```python
model.predict([[3, 22, 7 ,1]])
```

```python
y_predicted = model.predict(inputs_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(target_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
